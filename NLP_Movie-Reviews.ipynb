```python
from nltk.corpus import wordnet
def posTag(tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        else:
            return wordnet.NOUN
```

```python
documents = []
for i in category:
    for j in movie_reviews.fileids(i):
        documents.append((movie_reviews.words(j),i))
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('English'))
punctuations = set(string.punctuation)
stops = stops.union(punctuations)
```

```python
stops
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
pos_tag(['word'])
```

```python
from nltk import pos_tag
def clean_words(words):
    clean_words = []
    for i in words:
        #print(i)
        if i.lower() not in stops:
            #print(pos_tag([i]))
            pos = posTag(pos_tag([i])[0][1])   #pos_tag always takes list as input.
            clean_word = lemmatizer.lemmatize(i,pos = pos)
            clean_words.append(clean_word)
    return clean_words
```

```python
documents = [(clean_words(document),category) for document,category in documents]
```

For feature set to be used with nltk naive bayes classifier, we first create a list with all words from the training set. We then create a frequency object through the nltk library. The frequency object will then yield the most commonly occuring words which would then constitute the feature set.

Then we require for the classifier the training data in the form of tuples with a dictionary having true/false for features and the category.

```python
training_data = documents[0:1500]
testing_data = documents[1500:]
```

```python
all_Words = []
for i in training_data:
    all_Words += i[0]
len(all_Words)  
```

```python
from nltk import FreqDist
freqDist = FreqDist(all_Words)
```

```python
freqDist
```

```python
features = freqDist.most_common(3000)
features
```

```python
features = [feature[0] for feature in features]
```

```python
features
```

```python
def getDataInFormat(data):
    FormattedData = []
    for document,category in data:
        wordSet = set(document)
        FeatureDict = {}
        for feature in features:
            FeatureDict[feature] = feature in wordSet
        FormattedData.append((FeatureDict,category))
    return FormattedData
```

```python
TrainingData = getDataInFormat(training_data)
```

```python
TestingData = getDataInFormat(testing_data)
```

```python
TrainingData[0]
```

```python
count = 0
for key,value in TrainingData[0][0].items():
    if value == True:
        print(key)
        count+=1
print(count)
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(TrainingData)
```

```python
nltk.classify.accuracy(classifier,TestingData)
```

```python
classifier.show_most_informative_features(15)
```

## Using SKlearn Classifiers with NLTK Data

```python
from sklearn.svm import SVC
from nltk.classify.scikitlearn import SklearnClassifier ## NLTK has this dummy classifier to which if we pass SVC and format of data in tuples it will do everything for us automatically.
```

```python
svc = SVC(gamma = 'auto')
classifier_sklearn = SklearnClassifier(svc)
```

```python
classifier_sklearn.train(TrainingData)
```

```python
nltk.classify.accuracy(classifier_sklearn,TestingData)
```
